```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
```

```python
csvdata = pd.read_csv('Salary_Data.csv') 
```

```python
csvdata
```

```python
csvdata.describe()
```

```python
y = csvdata['Salary']
x1 = csvdata['YearsExperience']
```

```python
plt.scatter(x1,y)
plt.xlabel('Tahun',fontsize=40)
plt.ylabel('Gaji',fontsize=40)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
results
```

```python
plt.scatter(x1,y)
garis = 9449.9623*x1 + 2.579e+04
fig = plt.plot(x1,garis,lw=4,c='pink',label='garis linier regresi')
plt.xlabel('Tahun',fontsize=40)
plt.ylabel('Gaji',fontsize=40)
plt.show()
```
